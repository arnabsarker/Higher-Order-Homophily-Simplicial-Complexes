```python
import pandas as pd
```

```python
datasets = ['cont-village', 'cont-hospital', 'cont-workplace-13', 'cont-workplace-15', 
            'email-Enron', 'cont-primary-school', 'bills-senate', 'cont-high-school', 
            'bills-house', 'hosp-DAWN', 'coauth-dblp', 
            'clicks-trivago']

pred_datasets = ["cont-hospital", "cont-workplace-13", "cont-workplace-15", "hosp-DAWN", "bills-senate", 
                    "bills-house", "coauth-dblp", "cont-primary-school", "cont-high-school", 'email-Enron']
            
max_simplex_size = 8
```

```python
results_arr = []
for dataset in datasets:
    curr_info = {'dataset': dataset}
    label_df = pd.read_csv(f"../ScoreData/{dataset}/labels.csv")
    curr_info['nodes'] = len(label_df)
    curr_info['classes'] = len(pd.unique(label_df['group_code']))
    
    for k in range(2, max_simplex_size + 1):
        filename = f"../ScoreData/{dataset}/size-{k}-unique-sorted.csv"
        with open(filename, 'r') as filehandle:
            counter = -1
            for line in filehandle:
                counter += 1
        if(k == 2):
            curr_info['edges'] = counter
            if(dataset in pred_datasets):
                with open(f'../LinkPredictionData/{dataset}/times.txt', 'r') as file:
                    tlines = file.readlines()
                curr_info['timesteps'] = len(pd.unique(tlines))
            else:
                curr_info['timesteps'] = 0
        elif(k == 3):
            curr_info['triangles'] = counter
        elif(k == 4):
            curr_info['tetrahedra'] = counter
        elif(k > 4):
            curr_info[f'groups of size {k}'] = counter
    results_arr.append(curr_info)
```
